## 6. Which has the highest rating: TV shows or Movies?

```python
import pandas as pd

# Calculate the counts of each rating for Movies and TV Shows
rating_counts = df.groupby("type")["rating"].value_counts().unstack()

# Display the sum of ratings for each type
rating_counts.sum(axis=1)
```

```python
df.groupby("type")['rating'].value_counts().unstack().plot(kind = 'bar',figsize = (10,5),colormap ='viridis')
plt.title("Comparision of ratings for Movies vs TV shows")
plt.ylabel("Count")
plt.show()
```

```python
type_counts = df["type"].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(type_counts, labels=type_counts.keys(), autopct="%1.1f%%", colors=["#ffefa0", "#90e0ef"], startangle=90)
plt.title("Distribution of Movies vs. TV Shows")
plt.show()
```

## 9. Released movies over the years?

```python
sns.histplot(df[df["type"] == "Movie"]["release_year"], bins=30, kde=True, color = 'teal')
plt.title("Movies Released Over the Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Movies")
plt.grid()
plt.show()
```

## 10. Movies made on a yearly basis?

```python
print(df[df["type"] == "Movie"]["release_year"].value_counts().sort_index())
```

```python
plt.figure(figsize=(12, 5))
movies_per_year = df[df['type'] == 'Movie']['release_year'].value_counts().sort_index()

sns.barplot(x=movies_per_year.index, y=movies_per_year.values, hue=movies_per_year.index, palette="coolwarm", legend=False)
plt.xticks(rotation=90)
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.title("Movies Made on a Yearly Basis")
plt.show()
```
